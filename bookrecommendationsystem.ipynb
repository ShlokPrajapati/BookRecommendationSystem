# Popularity Based

```python
ratings_books=ratings.merge(books,on="ISBN")
ratings_books
```

```python
num_ratings=ratings_books.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_ratings.rename(columns={'Book-Rating':"num_ratings"},inplace=True)
num_ratings
```

```python
# Convert 'Book-Rating' column to numeric, coerce non-numeric values to NaN
ratings_books['Book-Rating'] = pd.to_numeric(ratings_books['Book-Rating'], errors='coerce')

# Drop rows with NaN values in 'Book-Rating' column
ratings_books.dropna(subset=['Book-Rating'], inplace=True)

# Calculate average ratings
avg_ratings = ratings_books.groupby("Book-Title")['Book-Rating'].mean().reset_index()

# Rename the column
avg_ratings.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

# Display the DataFrame
print(avg_ratings)
```

```python
popular_df=avg_ratings.merge(num_ratings,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
popular_df
```

# Collaborative Filtering Based Recommendation System


### Users who rated above 200 and books rated above and equal 50

```python
ratings_books
```

```python
x=ratings_books.groupby('User-ID').count()['Book-Rating']>200
Users_200=x[x].index
Users_200
```

```python
filtered_users=ratings_books[ratings_books['User-ID'].isin(Users_200)]
filtered_users
```

```python
y=filtered_users.groupby('Book-Title').count()['Book-Rating']>=50
users_200_books_50_ratings=y[y].index
```

```python
users_200_books_50_ratings
```

```python
final_filtered=filtered_users[filtered_users['Book-Title'].isin(users_200_books_50_ratings)]
final_filtered
```

```python
final_filtered.duplicated().sum()
```

```python
pivot_table=final_filtered.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pivot_table
```

```python
pivot_table.duplicated().sum()
```

```python
pivot_table.fillna(0,inplace=True)
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot_table)
# similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
def recommend_us(book_name):
    index=np.where(pivot_table.index==book_name)[0][0]
    distances=similarity_scores[index]
    books_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])
    recommended_books=[]
    for i in books_list[1:6]:
        individual_book=[]
        book_name=books[books['Book-Title']==pivot_table.index[i[0]]]
        individual_book.extend(list(book_name.drop_duplicates('Book-Title')['Book-Title'].values))
        individual_book.extend(list(book_name.drop_duplicates('Book-Title')['Book-Author'].values))
        individual_book.extend(list(book_name.drop_duplicates('Book-Title')['Image-URL-S'].values))
                        
        recommended_books.append(individual_book)
    return recommended_books
```

```python
recommend_us('You Belong To Me')
```

```python
recommend_us('The Da Vinci Code')
```
